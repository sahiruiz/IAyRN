```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('MachineLearning/Datasets/robot_sensors.csv')
print(df.head(10))
```

```python
print(df.isnull().sum())
```

```python
df_velneg = df[df['Velocidad (cm/s)'] >= 0]
df_temperatura = df[df['Temperatura (°C)'] >= 0]
print(df_velneg.head(10))
print(df_temperatura.head(10))
```

```python
def min_max_scaling(df_velneg):
    min_vals = np.min(df_velneg, axis=0)
    max_vals = np.max(df_velneg, axis=0)
    return (df_velneg - min_vals) / (max_vals - min_vals)
norm= min_max_scaling(df_velneg)
print("\nDatos normalizados:\n", norm)
```

```python

plt.figure(figsize=(15, 5))

plt.subplot(1, 2, 2)
for i in range(norm.shape[1]):
    plt.plot(norm.iloc[:, i], marker='.', linestyle='', label=f'Columna {i+1}')
plt.title("Datos Normalizados (Min-Max Scaling)")
plt.xlabel("Índice")
plt.ylabel("Valor Escalado")
plt.legend()

plt.tight_layout()
plt.show()
```

**1.¿Cómo afecta la limpieza de datos al análisis del dataset?**

La limpieza de datos es el proceso de corregir o eliminar datos incorrectos, duplicados o incompletos. Al aplicarla nos ayuda a obtener los datos con mayor claridad y sin que los datos se etiqueten incorrectamente.

**2.¿Por qué es importante la normalización en Machine Learning?**

Es importante porque permite crear normas o estándares que establezcan las características comúnes que se deben cumplir.
